# EDA with wine quality dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('C:/wine dataset.csv')
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
print(" First 5 Rows:\n", df.head())
print("\n Data Info:\n")
print(df.info())
print("\n Missing Values:\n", df.isnull().sum())
print("\n Summary Statistics:\n", df.describe())
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.set(style='whitegrid')
top_alcohols = df['alcohol'].round(1).value_counts().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_alcohols.index, y=top_alcohols.values, palette='Blues_d')
plt.title("Top 10 Alcohol Content Frequencies")
plt.xlabel("Alcohol % (rounded)")
plt.ylabel("Frequency")
plt.show()
```

```python
binned = pd.cut(df['alcohol'], bins=[10, 11, 12, 13, 14, 15], labels=['10-11','11-12','12-13','13-14','14-15'])
alcohol_pie = binned.value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(alcohol_pie, labels=alcohol_pie.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Alcohol Content Ranges (Pie Chart)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['alcohol'], df['proline'], alpha=0.7, color='purple')
plt.title("Alcohol vs Proline")
plt.xlabel("Alcohol")
plt.ylabel("Proline")
plt.grid(True)
plt.show()
```
